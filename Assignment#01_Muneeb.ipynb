# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
slice_data=data[['SL','SW','PL','PW']]
sns.heatmap(slice_data.corr(),annot=True)
plt.title("Correlation of SL,SW,PL,PW")
```

```python
sns.boxplot(x="Classification",y="PL",data=data)
plt.title("Relation Between Petal Lenghth and Iris Classification")
```

```python
sns.violinplot(x="Classification",y="PL",data=data)
plt.title("Relation Between Petal Lenghth and Iris Classification")
```

```python
sns.set_style("whitegrid")
sns.pairplot(data,hue="Classification",height=3)
plt.title("Pairplot")
```

```python
plt.figure(figsize=(6,4))
sns.swarmplot(x="Classification",y="SL",data=data,size=4)
plt.title("Relation Between Sepal Length and Iris Classification")
```
