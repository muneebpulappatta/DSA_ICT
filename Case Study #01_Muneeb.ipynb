# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,10))
sns.countplot('Buyer Gender',ec='black',data=data)
plt.title("COMPARISON OF MALE AND FEMALE BUYERS PARTICIPATED IN SALES",fontsize=16)
```

From the Bar Plot, We can Conclude Number of Female Buyers are greater than Number of Male Buyers.
